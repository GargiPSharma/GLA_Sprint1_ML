```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Student_id" : [101, 102, 103, 104, 105, 106],
    "Attendance_percentage" : [92, 67, 81, 45, 74, 88],
    "Assignment_score" : [18, 12, 15, 8, 14, 19],
    "Quiz_score" : [72, 48, 65, 30, 55, 80],
    "Lab_completed" : [True, False, True, False, True, True]
}

df = pd.DataFrame(data)
df
```

```python
df.shape
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
df["Total_score"] = df["Assignment_score"] + df["Quiz_score"]
df
```

```python
#Average Score
df["Average_score"] = df["Total_score"] / 2
df
```

```python
#Attendance Criterion
df["Attendance_Category"] = df["Attendance_percentage"].apply(
    lambda x: "High" if x >= 75 else "Low"
)

df
```

```python
#Performance
df["Performance"] = df["Average_score"].apply(
    lambda x: "Excellent" if x >= 70 else
              "Good" if x >= 50 else
              "Needs Improvement"
)

df
```

```python
#Score Percentage
df["Score_Percentage"] = (df["Total_score"] / 100) * 100

df
```

```python
#Result
df["Result"] = df["Average_score"].apply(
    lambda x: "Pass" if x >= 40 else "Fail"
)

df
```

```python
#Rank
df["Rank"] = df["Total_score"].rank(
    ascending=False,
    method="dense"
)

df
```

```python
#Visualization
```

```python
import matplotlib.pyplot as plt
```

```python
#Total Score Bar Graph
plt.figure(figsize=(8,5))

plt.bar(df["Student_id"].astype(str),
        df["Total_score"])

plt.title("Total Score of Students")
plt.xlabel("Student ID")
plt.ylabel("Total Score")

plt.show()
```

```python
#Average Score Line Graph
plt.figure(figsize=(8,5))

plt.plot(
    df["Student_id"],
    df["Average_score"],
    marker='o'
)

plt.title("Average Score of Students")
plt.xlabel("Student ID")
plt.ylabel("Average Score")

plt.grid(True)

plt.show()
```

```python
#Student Rank Bar Graph
rank_df = df.sort_values("Rank")

plt.figure(figsize=(8,5))

plt.bar(rank_df["Student_id"].astype(str),
        rank_df["Rank"])

plt.title("Student Ranking")
plt.xlabel("Student ID")
plt.ylabel("Rank")

plt.show()
```

```python
#Total Score vs Average Score
plt.figure(figsize=(8,5))

plt.plot(
    df["Student_id"],
    df["Total_score"],
    marker='o',
    label='Total Score'
)

plt.plot(
    df["Student_id"],
    df["Average_score"],
    marker='s',
    label='Average Score'
)

plt.title("Total Score vs Average Score")
plt.xlabel("Student ID")
plt.ylabel("Scores")

plt.legend()
plt.grid(True)

plt.show()
```

```python
#Performance Wise Average Score Bar Graph
performance_avg = df.groupby("Performance")["Average_score"].mean()

plt.figure(figsize=(8,5))

performance_avg.plot(kind="bar")

plt.title("Average Score by Performance Category")
plt.xlabel("Performance")
plt.ylabel("Average Score")

plt.show()
```
